## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#create data frame
schooltypes=school_data.set_index(["school_name"])["type"]

#create groupbys
school_counts=school_data_complete['school_name'].value_counts()
schoolbudget=school_data_complete.groupby(["school_name"]).mean()["budget"]
student_budget=schoolbudget/school_counts
school_math_score=school_data_complete.groupby(["school_name"]).mean()["math_score"]
school_reading_score=school_data_complete.groupby(["school_name"]).mean()["reading_score"]

school_passing_math=school_data_complete[(school_data_complete["math_score"]>=70)]
school_percentage_math=school_passing_math.groupby(["school_name"]).count()["student_name"]/school_counts*100

school_passing_reading=school_data_complete[(school_data_complete["reading_score"]>=70)]
school_percentage_reading=school_passing_reading.groupby(["school_name"]).count()["student_name"]/school_counts*100

school_passing_math_reading=school_data_complete[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70)]
school_percentage_math_reading=school_passing_math_reading.groupby(["school_name"]).count()["student_name"]/school_counts*100


summary=pd.DataFrame({"School Type": schooltypes,
                     "Total Students": school_counts,
                     "Total School Budget": schoolbudget,
                      "Per Student Budget": student_budget,
                     "Average Math Score":school_math_score,
                     "Average Reading Score":school_reading_score,
                     "% Passing Math": school_percentage_math,
                     "% Passing Reading": school_percentage_reading,
                     "% Overall Passing": school_percentage_math_reading})
summary
```
